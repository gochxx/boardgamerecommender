# API-Anfrage testen

```python
response = requests.get("https://api.geekdo.com/xmlapi/boardgame/161936?")
```

```python
root = ET.fromstring(response.content)
```

```python
def print_xml_structure(element, indent=0):
    print(' ' * indent + element.tag)
    for child in element:
        print_xml_structure(child, indent + 2)

print_xml_structure(root)
```

```python
# Über den XML-Tree iterieren und alle Elemente auslesen
result = {}
for boardgame in root.findall('.//boardgame'):
    for element in boardgame:
        if element.tag != "poll":
            # wenn das bereits in result, dann anfügen, sonst neu anlegen 
            if element.tag in result:
                result[element.tag].append(element.text)
            else:
                result[element.tag] = [element.text]
            print(element.tag, element.text)
        else:
                                    
            result[element.get('name')] = []
            for child in element:
                #print("Child-Elemente")
                #print(child.tag, child.attrib, child.text)
                
                for subchild in child:
                    #print("SubChild Elemente")
                    #print(subchild.tag, subchild.attrib, subchild.text)
                    result[element.get('name')].append(subchild.attrib)
```
